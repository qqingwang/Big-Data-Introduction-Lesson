## 1. 颜色
#### matplotlib 提供 8 种内建默认颜色:
- b: blue
- g: green
- r: red
- c: cyan
- m: magenta
- y: yellow
- b: black
- w: white 
#### 其他颜色表示方法:
- 灰色阴影：color='0.5'
- 十六进制：color='#D92090'
- RGB 形式：color=[0.1, 0.4, 0.6]

```python
#默认情况
y = np.arange(1,5)
plt.plot(y)
plt.title('Default color')
plt.show()

#调整颜色
y = np.arange(1,5)
plt.plot(y,'y');
plt.plot(y+1,color=(0.1,0.2,0.3));
plt.plot(y+2,'#FF00FF');
plt.plot(y+3,color='0.5')
plt.title('Different Colors')
plt.show()
```

## 2、点、线的样式、线型
#### 23 种点形
- ".": point
- ",": pixel
- "o": circle
- "v": triangle_down
- "^": triangle_up
- "<": triangle_left
- ">": triangle_right
- "1": tri_down
- "2": tri_up
- "3": tri_left
- "4": tri_right
- ......

#### 4 种线形
- -实线
- -- 虚线
- -. 点划线
- ：点线

#### 样式字符串
将颜色、点形、线形表达为字符串(颜色、点形、线形同时拼接为字符串)
- cx--
- mo:
- ko-

```python
#线型
y=np.arange(1,5)
plt.plot(y,'--');
plt.plot(y+1,'-.');
plt.plot(y+2,':');
plt.show()
```

```python
#点形状
y=np.arange(1,5)
plt.plot(y,'o');
plt.plot(y+1,'D');
plt.plot(y+2,'^');
plt.plot(y+3,'s');
plt.plot(y+4,'p');
plt.plot(y+5,'x');
plt.show()
```

```python
# 样式字符串形状
y=np.arange(1,5)
plt.plot(y,'cx--');
plt.plot(y+1,'kp:');
plt.plot(y+2,'mo-.');
plt.show()
```

## 3、matplotlib 支持三种画图方式
- pyplot：经典高级封装，使用方便但底层定制化不足
- pylab：将 matplotlib 和 numpy 合并，模拟 matlab 编程环境（不推荐使用）
- 面向对象的方式：支持更基础和底层的操作设置，难度较大

```python
#pyplot
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,10,1)
y = np.random.randn(len(x))
plt.plot(x,y)
plt.title('pyplot-random numbers')
plt.show()
```

```python
#pylab
from pylab import *
x = arange(0,10,1)
y = randn(len(x))
plot(x,y)
title('pylab-random numbers')
show()
```

```python
#Object Oriented
import matplotlib.pyplot as plt
import numpy as np

# 准备数据
x = np.arange(0,10,1)
y = np.random.randn(len(x))

'''
1. 生成画布对象
2. 在画布上生成坐标轴对象
3. 画图
'''
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x,y)
t = ax.set_title('object oriented-random numbers')
plt.show()
```

## 4. 子图 & 多图
Matplotlib 对象
- FigureCanvas
- Figure
- Axes

1. plt.figure() 构造画布实例对象
2. ax = fig.add_subplot(111) 在画布实例对象上添加坐标轴对象
- "111" 参数一：子图总行数
- "111" 参数二：子图总列数
- "111" 参数三：子图位置

```python
import matplotlib.pyplot as plt
import numpy as np

# 面向对象画子图
x=np.arange(1,100)

fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax1.plot(x,x)
ax2.plot(x,-x)
ax3.plot(x,x*x)
# ax4.plot(x,np.log(x))
plt.show()
```

```python
# pyplot画子图
x=np.arange(1,100)

plt.subplot(221)
plt.plot(x,x)

plt.subplot(222)
plt.plot(x,-x)

plt.subplot(223)
plt.plot(x,x*x)

plt.subplot(224)
plt.plot(x,np.log(x))

plt.show()
```

- 子图：一张图里生生多张子图
- 多图：直接生成多张图

```python
import matplotlib.pyplot as plt

# 多图（同一时间生成多张图）
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.set_title('Multi-graphs 1')
ax1.plot([1,2,3],[3,2,1])

fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
ax2.plot([1,2,3],[1,2,3])
ax2.set_title('Multi-graphs 2')
plt.show()
```

```python
# 调整日期
fig = plt.figure()
start = datetime.datetime(2015, 1, 1)
stop = datetime.datetime(2016, 1, 1)
delta = datetime.timedelta(days=1)

# 转化为 matplotlib理解的时间序列
dates = mpl.dates.drange(start, stop, delta)
y = np.random.rand(len(dates))

# 自定义日期格式
# date_format = mpl.dates.DateFormatter('%Y-%m-%d')
# ax = plt.gca()
# ax.xaxis.set_major_formatter(date_format)
# ax.plot_date(dates,y,linestyle='-',marker='')

plt.plot_date(dates,
              y,
              linestyle='-',
              marker='')

fig.autofmt_xdate()
plt.show()
```

#### 7.3 添加坐标轴

```python
import numpy as np
import matplotlib.pyplot as plt

# 添加一个坐标轴，形成双坐标轴
x = np.arange(2,20,1)

y1 = x*x
y2 = np.log(x)
plt.plot(x, y1)
plt.twinx() # 添加x坐标轴
plt.plot(x, y2, 'r')
plt.title('add x-axis')
plt.show()
```

```python
plt.plot(x, y1)
plt.twiny()
#plt.plot(x, y2,'r')
plt.plot(y2, x,'r')
plt.title('add y-axis')
plt.show()
```

```python
# 添加坐标轴（面向对象的方式）
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(x,y1)
ax1.set_ylabel('Y1')
ax2 = ax1.twinx()
ax2.plot(x,y2,'r')
ax2.set_ylabel('Y2')
ax1.set_xlabel('Compare Y1 and Y2')
plt.show()

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(x,y1)
ax1.set_xlabel('X1')
ax2 = ax1.twiny()
ax2.plot(y2,x,'r')
ax2.set_xlabel('X2')
ax1.set_ylabel('Compare X1 and X2')
plt.show()
```

## 8. 注释 & 文字 

```python
import numpy as np
import matplotlib.pyplot as plt

# 注释
x = np.arange(-10,11,1)
y = x*x
plt.plot(x,y)
plt.annotate('This is the bottom', #注释文字
             xy=(0, 1),     #注释形状的起始点（这里是箭头的坐标）
             xytext=(0,20), #注释文本的坐标
             # 注释形状
             arrowprops=dict(facecolor='r', #形状颜色
                             headlength=10, #箭头长度
                             headwidth=20,  #箭头宽度
                             width=10))     #箭头线段宽度
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 文字
x = np.arange(-10,11,1)
y = x*x
plt.plot(x,y)
# family 字体  style=normal/italic斜体  weight 字体粗细
plt.text(-1, #位置x坐标
         40, #位置y坐标
         'function:y=x*x', #文本
         family='serif',   #字体 serif/cursive/monospace
         size=20,
         color='r',
         style='italic',  #normal/italic/oblique
         weight='roman',  #light/normal/regulay/book/medium/roman/bold/haevy/black
         bbox=dict(facecolor='r', alpha=0.1))  # 箱体
plt.show()
```

## 9. Tex 公式
- matplotlib 自带 mathtext 引擎，不需要安装 TeX 系统
-   \$作为开始符和结束符

```python
# tex公式
fig = plt.figure()

ax = fig.add_subplot(111)
ax.set_xlim([1, 7])
ax.set_ylim([1, 5])

# r代表不转译
ax.text(2, 4, r"$\alpha_i \beta_i \pi \lambda \omega$", size=20)
ax.text(4, 4, r"$\sin(0)=\cos(\frac{\pi}{2})$", size=20)
ax.text(2, 2, r"$\lim_{x \rightarrow y}\frac{1}{x^3}$", size=20)
ax.text(4, 2, r"$\sqrt[4]{x}=\sqrt{y}$", size=20)

plt.show()
```

## 10. 区域填充

```python
import numpy as np
import matplotlib.pyplot as plt

# 区域填充
x = np.linspace(0,5*np.pi,1000)

y1 = np.sin(x)
y2 = np.sin(2*x)

# plt.plot(x,y1)
# plt.plot(x,y2)

plt.fill(x,y1,'b',alpha=0.3)
plt.fill(x,y2,'r',alpha=0.3)

fig = plt.figure()
ax = plt.gca()
ax.plot(x, y1, color='red')
ax.plot(x, y2, color='blue')
# 由于x的取值离散的，所以需要用到interpolate进行连续的填充
ax.fill_between(x,y1,y2,where=y1>y2,facecolors='yellow',interpolate=True)
ax.fill_between(x,y1,y2,where=y1<y2,facecolors='green',interpolate=True)
plt.show()
```
