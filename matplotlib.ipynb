```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

## 1. Setting Global Parameters

```python
# Set the global default size of matplotlib figures
plt.rc('figure', figsize=(10, 5))

# Set seaborn aesthetic parameters to defaults
seaborn.set()
```

## 2. Basic Plots

```python
# 0～2之间产生10个点
x = np.linspace(0, 2, 10)

plt.plot(x, x, 'o-', linestyle='--', linewidth=1.0, color='red',  label='linear')
plt.plot(x, x**2, 'x-', label='quadratic')

plt.legend(loc='best')
plt.title('Linear vs Quadratic progression')
plt.xlabel('Input')
plt.ylabel('Output');
plt.show()
```

## 3. Histograms

```python
# Gaussian, mean 1, stddev .5, 1000 elements
samples = np.random.normal(loc=1.0, scale=0.5, size=1000)
print(samples.shape)
print(samples.dtype)
print(samples[:30])
plt.hist(samples, bins=50);
plt.show()
```

## Two Histograms on the Same Plot

```python
samples_1 = np.random.normal(loc=1, scale=.5, size=10000)
samples_2 = np.random.standard_t(df=10, size=10000)
bins = np.linspace(-3, 3, 50)

# Set an alpha and use the same bins since we are plotting two hists
# x : 指定每个bin(箱子)分布的数据,对应x轴
# bins : 指定 bin (箱子)个数
# alpha 颜色深度

plt.hist(samples_1, bins=bins, alpha=0.5, label='samples 1')
plt.hist(samples_2, bins=bins, alpha=0.5, label='samples 2')
plt.legend(loc='upper left');
plt.show()
```

## 3. Scatter Plots

```python
samples_1 = np.random.normal(loc=1, scale=.5, size=10000)
samples_2 = np.random.standard_t(df=10, size=10000)

plt.scatter(samples_1, samples_2, alpha=0.1);
plt.show()
```

```python
plt.xlabel('X')
plt.ylabel('Y')

x1 = np.random.normal(2,1.2,300) # 随机产生300个平均值为2，方差为1.2的浮点数，即第一簇点的x轴坐标
y1 = np.random.normal(2,1.2,300) # 随机产生300个平均值为2，方差为1.2的浮点数，即第一簇点的y轴坐标
x2 = np.random.normal(7.5,1.2,300)
y2 = np.random.normal(7.5,1.2,300)  
area = np.pi * 4**2  # 点面积 

plt.scatter(x1, y1, s=area, c='#00CED1', alpha=0.4, label='Class A')
plt.scatter(x2, y2, s=area, c='#DC143C', alpha=0.4, label='Class')
plt.legend()

plt.show()
```
